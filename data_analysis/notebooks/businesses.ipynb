```python
import requests
import random

import pandas as pd
```

[Data source](https://data.colorado.gov/Business/Business-Entities-in-Colorado/4ykn-tg5h)

```python
# queries the socrata data api for Denver Nonprofits founded in the past 6 years
url = 'https://data.colorado.gov/resource/4ykn-tg5h.json?entitytype=Nonprofit Corporation&principalcity=Denver&$where=date_extract_y(entityformdate) > 2010&entitystatus=Good Standing'
res = requests.get(url)
df = pd.json_normalize(res.json())
df.head()
```
